# 🧩 Reto 4 – Integración final de resultados

**Objetivo:** combinar los resultados de los retos anteriores para obtener un diagnóstico integral de la calidad del agua.

### 🔹 Instrucciones
- Importa los resultados obtenidos en los notebooks anteriores.
- Integra los análisis (pH, temperatura, oxígeno) en un solo DataFrame.
- Realiza una conclusión general del estado del agua.

```python
import pandas as pd

# Cargar archivo
datos = pd.read_csv('../data/4_resultados_agua.csv')

# Ver  primeras filas
print("Primeras filas del archivo:")
print(datos.head())

# Ver nombres de columnas
print("\nColumnas del archivo:")
print(datos.columns)

# informacion 
print(datos.info())
```

```python
##Integracion de analisis

# Clasificación OD
datos["categoria_OD"] = "Normal"
datos.loc[datos["OD_promedio"] < 5, "categoria_OD"] = "Contaminada"


# Clasificación PH
def clasificar_ph(ph):
    if ph < 6.5:
        return "Ácido"
    elif ph <= 8.5:
        return "Neutro"
    else:
        return "Básico"

datos["categoria_ph"] = datos["pH"].apply(clasificar_ph)


# Clasificación TEMPERATURA
def clasificar_temp(temp):
    if temp < 20:
        return "Fría"
    elif temp <= 25:
        return "Media"
    else:
        return "Alta"

datos["categoria_temp"] = datos["Temp"].apply(clasificar_temp)


print("\nDatos integrados:")
print(datos)



```

```python
## Analisis general (promedios)

prom_ph = datos["pH"].mean()
prom_temp = datos["Temp"].mean()
prom_od = datos["OD_promedio"].mean()

print("\nPromedios:")
print("pH:", prom_ph)
print("Temperatura:", prom_temp)
print("Oxígeno:", prom_od)
```

```python
##CONCLUSION  
if prom_ph >= 6.5 and prom_od > 5:
    print("\nConclusión: El agua tiene buena calidad.")
elif prom_od < 5:
    print("\nConclusión: El agua presenta contaminación.")
else:
    print("\nConclusión: El agua tiene condiciones regulares.")
```

Se integraron los datos de pH, temperatura y oxígeno disuelto en un solo DataFrame para analizar la calidad del agua.
Se clasificaron los valores según condiciones ambientales dadas identificando posibles zonas con contaminacion.
Los resultados muestran que, en general, el agua presenta condiciones adecuadas, aunque algunos puntos pueden indicar afectaciones en la calidad.
