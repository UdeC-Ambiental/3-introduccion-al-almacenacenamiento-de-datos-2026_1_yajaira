# 📊 Reto 3 – Funciones y estadísticas en Pandas

**Objetivo:** aplicar funciones estadísticas y crear nuevas columnas basadas en cálculos.

```python
import pandas as pd
datos = pd.read_csv('../data/4_resultados_agua.csv')
```

### 🔹 Ejemplo de función personalizada
Crea una función que clasifique el pH:

```python
def clasificar_ph(valor):
    if valor < 6.5:
        return 'Ácido'
    elif valor <= 8.5:
        return 'Neutro'
    else:
        return 'Básico'

datos['categoria_ph'] = datos['pH'].apply(clasificar_ph)
datos.head()
```

### 💡 Actividad
- Calcula promedios de variables numéricas.
- Aplica al menos una función con `apply()` o `groupby()`.

```python
# Promedios generales
print("\nPromedios generales:")
print(datos.mean(numeric_only=True))
```

```python
# Promedios por categoría de pH
print("\nPromedios por categoría de pH:")
promedios_ph = datos.groupby('categoria_ph').mean(numeric_only=True)
print(promedios_ph)
```

```python
# Clasificación de temperatura
def clasificar_temp(temp):
    if temp < 20:
        return 'Fría'
    elif temp <= 25:
        return 'Media'
    else:
        return 'Alta'

datos['categoria_temp'] = datos['Temp'].apply(clasificar_temp)
```

```python
# Conteo por categoría de temperatura
print("\nCantidad por categoría de temperatura:")
print(datos['categoria_temp'].value_counts())
```

```python
print("\nInterpretación:")
print("Los datos muestran variaciones en temperatura y pH, lo que indica diferencias en las condiciones del agua.")
print("Las categorías permiten identificar comportamientos por grupo y facilitar el análisis.")
```

```python
print("\nConclusión:")
print("El uso de groupby y apply permite analizar los datos de manera más organizada, identificando patrones en la calidad del agua.")
```
