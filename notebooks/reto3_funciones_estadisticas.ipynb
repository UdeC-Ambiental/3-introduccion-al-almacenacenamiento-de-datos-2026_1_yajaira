### 🔹 Ejemplo de función personalizada
Crea una función que clasifique el pH:

```python
def clasificar_ph(valor):
    if valor < 6.5:
        return 'Ácido'
    elif valor <= 8.5:
        return 'Neutro'
    else:
        return 'Básico'

datos['categoria_ph'] = datos['pH'].apply(clasificar_ph)
datos.head()
```

```python
import pandas as pd

datos = pd.read_csv('../data/4_resultados_agua.csv')
```

```python
print("Promedios generales:")

print(datos.mean(numeric_only=True))
```

```python
datos['categoria_ph'] = ''

datos.loc[datos['pH'] < 7, 'categoria_ph'] = 'Ácido'
datos.loc[datos['pH'] == 7, 'categoria_ph'] = 'Neutro'
datos.loc[datos['pH'] > 7, 'categoria_ph'] = 'Básico'
```

```python
datos['categoria_temp'] = ''

# Primero asignas todos como 'Media'
datos.loc[datos['Temp'] >= 20, 'categoria_temp'] = 'Media'

# Luego corriges los menores
datos.loc[datos['Temp'] < 20, 'categoria_temp'] = 'Fría'

# Y por último los mayores
datos.loc[datos['Temp'] > 25, 'categoria_temp'] = 'Alta'
```

```python
print("\nPromedios por categoría de pH:")
print(datos.groupby('categoria_ph').mean(numeric_only=True))
```

```python
print("Cantidad por categoría de pH:")

print(datos['categoria_ph'].value_counts())
```
